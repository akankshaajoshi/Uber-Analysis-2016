# Uber data analysis

```python
#Importing required libraries

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#Reading csv

uber = pd.read_csv('uber16.csv')
```

```python
uber
```

```python
#Dropping rows with missing column values

uber = uber.dropna()
uber
```

```python
#Fixing typographic errors

uber = uber.replace({"Kar?chi": 'Karachi'})
uber
```

```python
#Basic information

uber.info()
```

```python
#Renaming columns

uber.rename(columns = {'START_DATE*': "START_DATE", 'END_DATE*': 'END_DATE', 'CATEGORY*': 'CATEGORY', 'START*': 'START', 'STOP*': 'STOP', 'MILES*': 'MILES', 'PURPOSE*': 'PURPOSE'}, inplace=True)
```

```python
uber
```

```python
#Conversion to datetime type columns

uber['START_DATE'] = pd.to_datetime(uber['START_DATE'])
uber['END_DATE'] = pd.to_datetime(uber['END_DATE'])
uber
```

```python
#Extracting more information

uber['HOUR'] = [x.hour for x in uber['START_DATE']]
uber['MINUTES'] = [x.minute for x in uber['START_DATE']]
uber['DAY'] = [x.day for x in uber['START_DATE']]
uber['MONTH'] = [x.month for x in uber['START_DATE']]
uber['MONTH'] = uber['MONTH'].replace({1:'January', 2:'February', 3:'March', 4:'April', 5: 'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11: 'November', 12: 'December'})
uber
```

## Analysis

```python
#Category vs frequency

sns.set_style('dark')
cat_count = pd.DataFrame(uber['CATEGORY'].value_counts())
cat_count
```

```python
plt.bar(cat_count.index, cat_count['CATEGORY'], color = ['blue', 'red'])
plt.show()

#Availed mainly for business use.
```

```python
#Purpose vs frequency

purpose = pd.DataFrame(uber['PURPOSE'].value_counts())
cmap = plt.get_cmap('viridis')
cmap_c  = cmap(purpose['PURPOSE'])
plt.barh(purpose.index, purpose['PURPOSE'], color=cmap_c)
plt.show()

#Major purpose for hiring a uber ride was to attend meetings/meals at work/work errands/customer visits.
#Users are primarilly working in client based companies which requries frequent reloaction and visits during the job.
```

```python
#Frequency plot for Number of miles travelled

sns.histplot(uber['MILES'])
plt.show()

#Right skewed normal distribution of number of miles travelled per ride.
```

```python
#Number of trips per hour

hours = uber['HOUR'].value_counts()
hours.plot(kind = 'bar', color='blue')
plt.xlabel('Hours')
plt.xticks(rotation=360)
plt.ylabel('Frequency')
plt.title('Hours vs Frequency')

#Around 11 A.M. - 6 P.M. maximum users availed uber rides.
```

```python
#Month vs frequency distribution

months = pd.DataFrame(uber['MONTH'].value_counts())
plt.pie(months['MONTH'], labels = months.index, startangle=0)
plt.show()

#In december 2016, uber rides increased significantly in number.
```

```python
#Common pickup locations

pickup = pd.DataFrame(uber['START'].value_counts())
plt.stem(pickup.index[0:5], pickup['START'][0:5])
plt.yticks([])
plt.show()
```

```python
#Common end locations

drop = pd.DataFrame(uber['STOP'].value_counts())
plt.stem(drop.index[0:5], drop['STOP'][0:5])
plt.yticks([])
plt.show()
```
